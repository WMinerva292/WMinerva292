### **Merging the DataFrames**

```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
#pd.merge(market, customer, how = 'inner', left_on = 'Cust_id', right_on = 'Id')
```

```python
new_df = pd.merge(market, customer, how = 'inner', on = 'Cust_id')
new_df.head()
```

```python
final_df = pd.merge(new_df, product, how = 'inner', on = 'Prod_id')
final_df.head()
```

```python
#for jupyter notebook users, try this
final_df.to_csv('merged_data.csv', index = False)
```

```python
final_df.to_excel('merged_data.xlsx', index = False)
```

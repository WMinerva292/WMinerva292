```python
# 10. Use a for loop to calculate the number of customers that are getting the tech support and are male senior citizens.

# Initialize counter
count = 0

# Loop through the dataset
for index, row in churn.iterrows():
    if row['gender'] == 'Male' and row['SeniorCitizen'] == 1 and row['TechSupport'] == 'Yes':
        count += 1

# Display the result
print(f"Number of male senior citizens with tech support: {count}")
```

```python
# 1. Write a function that takes start and end of a range returns a pandas series object containing numbers within that range.
# In case the user does not pass start or end or both they should default to 1 and 10 respectively. E.g:
# range_series() -> Should Return a pandas series from 1 to 10
# range_series(5) -> Should Return a pandas series from 5 to 10
# range_series(5, 10) -> Should Return a pandas series from 5 to 15
# Create a method that takes n NumPy arrays of the same dimensions, sums them and returns the answer.

import numpy as np
import pandas as pd

def range_series(start=1, end=10):
  return pd.Series(range(start, end+1))

# example function call
print("Series from 1 to 10:")
print(range_series())

print("\nSeries from 5 to 10:")
print(range_series(5))

print("\nSeries from 5 to 15:")
print(range_series(5,10))
```

```python
# Create a method that takes n NumPy arrays of the same dimensions, sums them and returns the answer.

import numpy as np

def sum_arrays(*arrays):
  return np.array(arrays).sum()
```

```python
# 2. Create a function that takes in two lists named keys and values as arguments
# Keys would be strings and contain n string values
# Values would be a list containing n lists
# The methods should return a new pandas DataFrame with keys as column names and values as their corresponding values, e.g:
# create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]]) -> should return a data frame
# One Two
# 0 X A
# 1 Y B

import numpy as np
import pandas as pd

def create_dataframe(keys, values):
  data = list(map(list, zip(*values)))
  return pd.DataFrame(data, columns = keys)

create_dataframe(["One","Two"],[["X","Y"],["A","B"]])
```

```python
# 3. Create a function that concatenates two DataFrames. Use a previously
# created function to create two DataFrames and pass them as parameters
# Make sure that the indexes are reset before returning.

def concatenate_dataframes(df1, df2):
  concatenated_df = pd.concat([df1, df2], axis = 0) .reset_index(drop=True)
  return concatenated_df

df1 = create_dataframe(["One","Two"],[["X","Y"],["A","B"]])
df2 = create_dataframe(["One","Two"],[["M","N"],["C","D"]])
concat_df = concatenate_dataframes(df1, df2)
#print("\nConcatenate DataFrames:")
print(concat_df)
```

```python
# 4. Write code to load data from cars.csv into a dataframe and print its details.
# Details like: 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'.

# Load the important libraries
import pandas as pd
import numpy as np

# Load the cars.csv dataset into a dataframe
df = pd.read_csv('/content/cars.csv')

# print its details
print("\nCars Dataframe Details:")
print(df.describe())
```

```python
# 5. Write a method that will take a column name as argument and return the name of the column
# with which the given column has the highest correlation.

#The data to be used is the cars dataset.

# The returned value should not the column named that was passed as the parameters,
# e.g. : get_max_correlated_column('mpg') -> should return 'drat

def get_max_correlated_column(df, column_name):
  correlation = df.corr(numeric_only=True)

  max_corr_col = correlation[column_name].drop(column_name).idxmax()
  return max_corr_col

max_corr = get_max_correlated_column(df, 'mpg')
print(f"\nColumn with highest correlation to 'mpg': {max_corr}")
```

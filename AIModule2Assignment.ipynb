**Introduction:**

Neural networks, primarily known for their applications in classification tasks, can also be employed for regression problems. In this exercise, we aim to
demonstrate the usage of a simple neural network in predicting a linear
relationship between input features and the output.

**Objective:**

The primary objective is to predict the output value y based on two input features x1 and x2 using a basic neural network model with the equation y=3x1+4x2.
The neural network will be trained using a feed-forward and back-propagation
mechanism to minimize the error between the predicted and actual values.

**Methodology:**

**1. Activation Functions:**

● Define the sigmoid function, a popular activation function used in neural
networks.

● Define the tanh function, which is another common activation function that
returns values between -1 and 1.

● Define the relu function, which returns positive values as they are and
zeroes out negative values.

**2. Sample Data:**

● Use the following sample data for training:

x1=3, x2=2, yactual=17

**3. Feed Forward:**

● Initialize the random weights w1 and w2 between 1 and 10.

● Predict the output ypred using the equation ypred= x1 x w1 + x2 x w2

**4. Error Calculation:**

● Compute the squared error as error = (yactual-ypred)*(yactual-ypred)

**5. Back Propagation:**

● Compute the gradients of the error with respect to the weights.

● Update the weights using the computed gradients and a learning rate.

**6. Training:**

● The above steps (Feed Forward to Back Propagation) are performed
iteratively (20 times in this example) to refine the weights and minimize
the error.

**7. Visualization:**

● Plot the progression of the error across epochs.

● Plot the progression of ypred across epochs.

**Expected Outcome:**

After sufficient training, the predicted ypred should be close to the actual value yactual=17. The error graph should show a declining trend, indicating that the model is learning and improving its predictions over time.

**Importing Libraries:**

```python
 # Import Libraries

 import numpy as np
 import matplotlib.pyplot as plt
```

**1. Activation Functions:**



```python
# Define the sigmoid function, a popular activation function used in neural networks.

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Define the tanh function, which is another common activation function that returns values between -1 and 1.

def tanh(x):
    return np.tanh(x)
```

```python
# Define the relu function, which returns positive values as they are and zeroes out negative values.

def relu(x):
    return np.maximum(0, x)
```

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  Use the following sample data for training: x1=3, x2=2, yactual=17
x1, x2 = 3, 2
y_actual = 17

# Initialize the random weights w1 and w2 between 1 and 10.
np.random.seed(42)
w1, w2 = np.random.uniform(1, 10, 2)
learning_rate = 0.01

epochs = 20
errors = []
predictions = []

for epoch in range(epochs):
    # Feed Forward: Predict the output ypred using the equation ypred= x1 x w1 + x2 x w2
    y_pred = x1 * w1 + x2 * w2

    # Error Calculation: Compute the squared error as error = (yactual-ypred)*(yactual-ypred)
    error = (y_actual - y_pred) ** 2
    errors.append(error)
    predictions.append(y_pred)

    # Compute the Gradients of the error with respect to the weights.
    grad_w1 = -2 * (y_actual - y_pred) * x1
    grad_w2 = -2 * (y_actual - y_pred) * x2

    # Update the Weights using the computed gradients and a learning rate.
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

    print(f"Epoch {epoch+1}: y_pred = {y_pred:.4f}, error = {error:.4f}")

# Visualization:
# Plot the progression of the error across epochs.
# Plot the progression of ypred across epochs.

    import matplotlib.pyplot as plt
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(list(range(len(errors))), errors, marker='o', label='Error')
    plt.xlabel('Epochs')
    plt.ylabel('Squared Error')
    plt.title('Error Reduction Over Epochs')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(list(range(len(predictions))), predictions, marker='s', label='Predicted y')
    plt.axhline(y=y_actual, color='r', linestyle='--', label='Actual y')
    plt.xlabel('Epochs')
    plt.ylabel('Predicted y')
    plt.title('Predicted y Progression')
    plt.legend()
    plt.show()
```

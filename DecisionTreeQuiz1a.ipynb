```python
# drop the missing values
df.dropna(inplace=True)
```

```python
# Split the dataset into a 70:30 ratio with random state as 1.
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Legendary', 'Name'] axis=1), df['Legendary'], test_size=0.3, random_state=1)
```

```python
# What is the precision of the Decision Tree model when the target is False?
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, pos_label=False)
print(precision)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, precision_score

# Load the dataset
# Replace with the actual path of your dataset
df = pd.read_csv('/content/Pokemon.csv')

# Handle missing values (Example: Fill NaN with the mean of each column)
df.fillna(df.mean(), inplace=True)

# Define features (X) and target (y)
X = df.drop(columns=["Legendary"])
y = df["Legendary"]

# Split the data into training and testing sets (70:30 ratio, random state = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Build the Decision Tree model
model = DecisionTreeClassifier(random_state=1)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate precision for when the target is False
precision_false = precision_score(y_test, y_pred, pos_label=False)

# Print the precision score
print(f"Precision for False Targets: {precision_false:.2f}")
```
